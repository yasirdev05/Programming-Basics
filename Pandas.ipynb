## `iloc` — Integer Location (Position-based)
Use when you know the **row/column number** (0, 1, 2...)
```python
df.iloc[row_position]
df.iloc[start:end]        # end is EXCLUSIVE
df.iloc[row, col]
```

## `loc` — Label Location (Name-based)
Use when you know the **row/column name or label**
```python
df.loc[row_label]
df.loc[start:end]         # end is INCLUSIVE
df.loc[row, 'ColName']
```

# 📌 Topic 5 — Filtering Rows (Boolean Indexing)

Filter rows by writing a **condition** inside `df[ ]`

```python
df[df['column'] > value]           # greater than
df[df['column'] == 'text']         # equals
df[condition1 & condition2]        # AND
df[condition1 | condition2]        # OR
df[df['col'].isin(['a', 'b'])]     # value in list
```
> ⚠️ Use `&` (not `and`) and `|` (not `or`) in Pandas conditions.  
> Wrap each condition in parentheses `( )`

````python
# ───────────────────────────────────────────
# ✅ EXAMPLE — Filtering / Boolean Indexing
# ───────────────────────────────────────────

print('── Students with CGPA > 3.5 ─────────')
print(df[df['CGPA'] > 3.5])

print('\n── Students from Lahore ─────────────')
print(df[df['City'] == 'Lahore'])

print('\n── CGPA > 3.5 AND Age < 23 ──────────')
print(df[(df['CGPA'] > 3.5) & (df['Age'] < 23)])

print('\n── From Lahore OR CGPA > 3.8 ────────')
print(df[(df['City'] == 'Lahore') | (df['CGPA'] > 3.8)])

print('\n── isin(): from Lahore or Karachi ───')
print(df[df['City'].isin(['Lahore', 'Karachi'])])
````

# 📌 Topic 9 — Handling Missing Values

Missing values appear as **NaN** (Not a Number). Pandas tools:

```python
df.isnull()          # True where value is NaN
df.isnull().sum()    # count of NaN per column
df.dropna()          # drop rows with any NaN
df.fillna(value)     # replace NaN with value
df['col'].fillna(df['col'].mean())   # fill with column mean
```

````python
# ───────────────────────────────────────────
# ✅ EXAMPLE — Handling Missing Values
# ───────────────────────────────────────────

# Create a DataFrame with NaN values
data_nan = {
    'Name' : ['Ali', 'Sara', None, 'Zara', 'Bilal'],
    'Score': [85, np.nan, 90, np.nan, 70],
    'Grade': ['A', 'A+', 'A', None, 'B+']
}
df_nan = pd.DataFrame(data_nan)

print('── Original DataFrame ───────────────')
print(df_nan)

print('\n── isnull() ─────────────────────────')
print(df_nan.isnull())

print('\n── Count of NaN per column ──────────')
print(df_nan.isnull().sum())

print('\n── After dropna() ───────────────────')
print(df_nan.dropna())

print('\n── Fill Score NaN with mean ─────────')
df_nan['Score'] = df_nan['Score'].fillna(df_nan['Score'].mean())
print(df_nan)
````

# 📌 Topic 10 — apply() with Custom Functions

**`apply()`** runs a function on every element in a column.

```python
# Named function
def my_func(x):
    return x * 2
df['NewCol'] = df['Col'].apply(my_func)

# Lambda (one-line anonymous function)
df['NewCol'] = df['Col'].apply(lambda x: x * 2)
```

````python
# ───────────────────────────────────────────
# ✅ EXAMPLE — apply()
# ───────────────────────────────────────────

df_apply = df.copy()

# Named function
def grade_from_cgpa(cgpa):
    if cgpa >= 3.7:
        return 'A'
    elif cgpa >= 3.3:
        return 'B+'
    else:
        return 'B'

df_apply['LetterGrade'] = df_apply['CGPA'].apply(grade_from_cgpa)
print('── Grade using named function ───────')
print(df_apply[['Name', 'CGPA', 'LetterGrade']])

# Lambda function
df_apply['Senior'] = df_apply['Age'].apply(lambda x: 'Yes' if x >= 22 else 'No')
print('\n── Senior flag using lambda ─────────')
print(df_apply[['Name', 'Age', 'Senior']])
````

# 📌 Topic 11 — groupby() and Aggregation

**`groupby()`** splits data into groups and lets you compute statistics per group.

```python
df.groupby('GroupCol')['ValueCol'].mean()    # mean per group
df.groupby('GroupCol').agg(                  # multiple aggregations
    Avg = ('col', 'mean'),
    Total = ('col', 'sum'),
    Count = ('col', 'count')
)
```

````python
# ───────────────────────────────────────────
# ✅ EXAMPLE — groupby()
# ───────────────────────────────────────────

print('── Average CGPA per City ────────────')
print(df.groupby('City')['CGPA'].mean())

print('\n── Count of students per City ───────')
print(df.groupby('City')['Name'].count())

print('\n── Multiple aggregations ────────────')
summary = df.groupby('City').agg(
    Avg_CGPA = ('CGPA', 'mean'),
    Count    = ('Name', 'count'),
    Max_Age  = ('Age',  'max')
)
print(summary)

# print('\n── Count by Enrolled status ─────────')
# print(df.groupby('Enrolled')['Name'].count())
````

# 📌 Topic 12 — merge() and concat()

**`concat()`** — Stack DataFrames vertically (add rows) or horizontally (add columns)
```python
pd.concat([df1, df2], ignore_index=True)   # stack rows
```

**`merge()`** — Join DataFrames on a common column (like SQL JOIN)
```python
pd.merge(df1, df2, on='CommonColumn')      # inner join
pd.merge(df1, df2, on='Col', how='left')   # left join
```

````python
# ───────────────────────────────────────────
# ✅ EXAMPLE — concat() and merge()
# ───────────────────────────────────────────

batch1 = pd.DataFrame({'Name': ['Ali', 'Sara'],  'CGPA': [3.5, 3.8]})
batch2 = pd.DataFrame({'Name': ['Ahmed', 'Zara'], 'CGPA': [3.2, 3.9]})

print('── concat(): stack rows ─────────────')
all_students = pd.concat([batch1, batch2], ignore_index=True)
print(all_students)

# merge example
df_info    = pd.DataFrame({'Name': ['Ali','Sara','Ahmed'], 'CGPA': [3.5, 3.8, 3.2]})
df_contact = pd.DataFrame({'Name': ['Ali','Sara','Ahmed'], 'Email': ['ali@umt.edu','sara@umt.edu','ahmed@umt.edu']})

print('\n── merge(): join on Name ────────────')
merged = pd.merge(df_info, df_contact, on='Name')
print(merged)
````
