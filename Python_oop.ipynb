## ✅ CHAPTER 1: What is a Class?

### 🔷 Definition
A **class** is a **blueprint/template** for creating objects.
Think of it like a **cookie cutter** 🍪 — the cutter is the class, and each cookie you make is an object.

### 🤔 Why Do We Create a Class?
Without classes, if you want to store data about 100 students, you'd need 100 × 3 = 300 variables!

```
# Without class — messy and unscalable 😬
student1_name = "Ali"
student1_age  = 20
student1_grade = "A"

student2_name = "Sara"
student2_age  = 19
student2_grade = "B"
# ... imagine doing this for 100 students!
```

With a class, you define the structure **once** and create as many objects as you need. ✅

### 📐 Syntax
```python

    # attributes (data)
    # methods (functions)
```

## ✅ CHAPTER 2: Attributes — Data Inside a Class

### 🔷 What is an Attribute?
An **attribute** is a **variable that belongs to a class or object**.
It stores the **data/state** of an object.

| Type | Where defined | Shared? |
|------|--------------|---------|
| **Class Attribute** | Inside class, outside methods | ✅ Shared by all objects |
| **Instance Attribute** | Inside `__init__` using `self` | ❌ Unique to each object |

---
### 📌 Class Attributes (Shared)


````python
# Class Attribute — shared by ALL objects
class Student:
    school = "City High School"   # ← Class attribute

s1 = Student()
s2 = Student()

print(s1.school)   # City High School
print(s2.school)   # City High School

# Changing via class changes for ALL
s1.school = "National Academy"
print(s1.school)   # National Academy ← changed!
print(s2.school)   # National Academy ← changed too!
````

## ✅ CHAPTER 3: Methods — Functions Inside a Class

### 🔷 What is a Method?
A **method** is a **function defined inside a class**.
It defines the **behavior** of the object — what it can **do**.

```
Class = Data (attributes) + Behavior (methods)
```

### 📌 Types of Methods

| Type | Decorator | First Param | Access |
|------|-----------|-------------|--------|
| Instance Method | None | `self` | Instance attributes |
| Class Method | `@classmethod` | `cls` | Class attributes |
| Static Method | `@staticmethod` | None | Independent |


````python
# Instance Methods — most common type
class Student:
    def __init__(self):
        self.name  = ""
        self.marks = 0

    def set_data(self, name, marks):     # ← setter method
        self.name  = name
        self.marks = marks

    def show_data(self):                 # ← getter/display method
        print(f"Name : {self.name}")
        print(f"Marks: {self.marks}")

    def get_grade(self):                 # ← logic method
        if self.marks >= 90:
            return "A+"
        elif self.marks >= 80:
            return "A"
        elif self.marks >= 70:
            return "B"
        else:
            return "C or below"

# Using the methods
s1 = Student()
s1.set_data("Ali", 88)
s1.show_data()
print(f"Grade: {s1.get_grade()}")
````

````python
# Class Method and Static Method
class MathHelper:
    pi = 3.14159           # class attribute

    @classmethod
    def circle_area(cls, radius):       # uses class attribute
        return cls.pi * radius * radius

    @staticmethod
    def add(a, b):                      # independent — no self or cls
        return a + b

    @staticmethod
    def is_even(n):
        return n % 2 == 0

# Calling without creating an object
print(MathHelper.circle_area(5))
print(MathHelper.add(10, 20))
print(MathHelper.is_even(4))
````

## ✅ CHAPTER 4: The `self` Keyword

### 🔷 What is `self`?
`self` refers to the **current object** that is calling the method.
It's how the object talks to **its own** attributes and methods.

```
Think of self as saying "MY name", "MY age", "MY marks"
instead of just "name", "age", "marks"
```

### 🔍 What happens without `self`?


````python
# Without self — BROKEN ❌
class Student:
    def set_name(self, name):
        name = name          # This creates a LOCAL variable, NOT stored in object!

    def show_name(self):
        print(name)          # ERROR — 'name' doesn't exist here

# With self — CORRECT ✅
class Student:
    def set_name(self, name):
        self.name = name     # Stored IN the object

    def show_name(self):
        print(self.name)     # Reads FROM the object

s = Student()
s.set_name("Hamza")
s.show_name()                # Output: Hamza
````

````python
# self refers to WHICHEVER object calls the method
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says: Woof!")   # self.name = THIS dog's name

d1 = Dog("Bruno")
d2 = Dog("Max")

d1.bark()   # Bruno says: Woof!   ← self = d1
d2.bark()   # Max says: Woof!     ← self = d2
````

## ✅ CHAPTER 5: Constructors (`__init__`)

### 🔷 What is a Constructor?
A **constructor** is a **special method** that runs **automatically** when an object is created.
In Python it is always named `__init__`.

```
Object created  →  __init__ runs automatically  →  Object is ready
```

### 🔑 Why use a Constructor?
Without a constructor you'd have to call `set_data()` manually every time.
The constructor lets you **set up the object instantly at creation**.


````python
# Without constructor — manual setup needed 😕
class Student:
    def set_data(self, name, age):
        self.name = name
        self.age  = age

s = Student()
s.set_data("Ali", 20)     # Must remember to call this!

# With constructor — automatic setup ✅
class Student:
    def __init__(self, name, age):
        self.name = name   # runs the MOMENT you write Student("Ali", 20)
        self.age  = age
        print(f"Student '{name}' created!")

s = Student("Ali", 20)    # Constructor runs here
print(s.name)
print(s.age)
````

### 🔍 How Python calls `__init__` behind the scenes
```
Student("Ali", 20)
    ↓
Python creates blank object
    ↓
Python calls __init__(new_object, "Ali", 20)
    ↓  (new_object becomes 'self')
self.name = "Ali"
self.age  = 20
    ↓
Object returned to you
```

## ✅ CHAPTER 6: Types of Constructors

### 🔷 Three Types

| Type | Parameters | Use when |
|------|-----------|----------|
| **Default** | Only `self` | Object needs no input at creation |
| **Parameterized** | `self` + extra | Object needs data at creation |
| **Default Arguments** | `self` + defaults | Object may or may not receive data |


````python
# TYPE 1: Default Constructor
class Circle:
    def __init__(self):           # no parameters
        self.radius = 1           # default value hardcoded
        self.color  = "red"

    def area(self):
        return 3.14 * self.radius ** 2

c = Circle()                      # no arguments needed
print(f"Radius: {c.radius}, Area: {c.area():.2f}")
````

````python
# TYPE 2: Parameterized Constructor
class Rectangle:
    def __init__(self, length, width):     # requires arguments
        self.length = length
        self.width  = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r1 = Rectangle(10, 5)
r2 = Rectangle(7, 3)

print(f"R1 — Area: {r1.area()}, Perimeter: {r1.perimeter()}")
print(f"R2 — Area: {r2.area()}, Perimeter: {r2.perimeter()}")
````

````python
# TYPE 3: Default Argument Constructor (most flexible)
class Employee:
    company = "TechCorp"           # class attribute

    def __init__(self, name="Unknown", salary=30000, dept="General"):
        self.name   = name
        self.salary = salary
        self.dept   = dept

    def show(self):
        print(f"{self.name} | {self.dept} | ${self.salary:,} | {Employee.company}")

# All three ways to create
e1 = Employee()                              # all defaults
e2 = Employee("Sara")                        # only name
e3 = Employee("Hamza", 55000, "Engineering") # all provided

e1.show()
e2.show()
e3.show()
````

## ✅ CHAPTER 7: Destructor (`__del__`)

### 🔷 What is a Destructor?
A **destructor** is a special method called **automatically when an object is destroyed**.
It is defined as `__del__`.

```
Object created  →  lives in memory  →  deleted/goes out of scope  →  __del__ runs
```

### 🔑 When is it used?
- Closing file connections
- Releasing database connections
- Cleanup tasks
- Logging/tracking object lifecycle


````python
# Basic Destructor example
class FileManager:
    def __init__(self, filename):
        self.filename = filename
        print(f"📂 File '{filename}' opened.")

    def read(self):
        print(f"📖 Reading from '{self.filename}'...")

    def __del__(self):
        print(f"🔒 File '{self.filename}' closed. (Destructor called)")

# Object lifecycle
fm = FileManager("data.txt")
fm.read()

print("--- deleting object now ---")
del fm    # Destructor called here

print("--- program continues ---")
````

````python
# Destructor called automatically when variable goes out of scope
class Tracker:
    count = 0   # tracks how many objects exist

    def __init__(self, name):
        self.name = name
        Tracker.count += 1
        print(f"✅ '{name}' created. Total objects: {Tracker.count}")

    def __del__(self):
        Tracker.count -= 1
        print(f"❌ '{self.name}' destroyed. Total objects: {Tracker.count}")

t1 = Tracker("Alpha")
t2 = Tracker("Beta")
t3 = Tracker("Gamma")

del t2   # Only Beta destroyed
print(f"Still active: {Tracker.count}")
````

## ✅ CHAPTER 8: Access Modifiers

### 🔷 What are Access Modifiers?
Access modifiers control **who can access** the attributes and methods of a class.

| Modifier | Syntax | Access |
|----------|--------|--------|
| **Public** | `self.name` | Anywhere ✅ |
| **Protected** | `self._name` | Class + subclasses (by convention) ⚠️ |
| **Private** | `self.__name` | Class only (name mangling) 🔒 |

> Python doesn't enforce these strictly — it's mostly a **convention** between developers.


````python
# All three access levels
class BankAccount:
    bank_name = "SafeBank"         # public class attribute

    def __init__(self, owner, balance, pin):
        self.owner    = owner       # public    — anyone can see
        self._balance = balance     # protected — "please don't touch from outside"
        self.__pin    = pin         # private   — name-mangled, hidden

    def show_balance(self):         # public method
        print(f"Owner  : {self.owner}")
        print(f"Balance: ${self._balance:,.2f}")

    def _calculate_interest(self):  # protected method
        return self._balance * 0.05

    def __verify_pin(self, pin):    # private method
        return self.__pin == pin

    def withdraw(self, amount, pin):
        if self.__verify_pin(pin):  # can call private method INSIDE class
            if amount <= self._balance:
                self._balance -= amount
                print(f"Withdrawn ${amount}. New balance: ${self._balance:,.2f}")
            else:
                print("Insufficient balance!")
        else:
            print("❌ Wrong PIN!")

acc = BankAccount("Ali", 5000, 1234)
acc.show_balance()

acc.withdraw(500, 1234)    # ✅ correct pin
acc.withdraw(200, 9999)    # ❌ wrong pin

# Public — accessible directly
print(acc.owner)

# Protected — accessible but shouldn't be
print(acc._balance)        # Works but ⚠️ bad practice

# Private — name mangled — direct access raises error
# print(acc.__pin)         # ← AttributeError!
print(acc._BankAccount__pin)  # Name mangling trick (not recommended)
````

## 📊 OOP Concepts Summary

```
OOP
├── CLASS          → Blueprint / Template
│   ├── Attributes → Data / State (what it HAS)
│   └── Methods    → Behavior (what it DOES)
│
├── OBJECT         → Instance of a class
│
├── CONSTRUCTOR    → __init__() auto-runs at object creation
│   ├── Default    → no params
│   ├── Parameterized → requires args
│   └── Default Args  → optional args with defaults
│
├── DESTRUCTOR     → __del__() auto-runs at object destruction
│
├── self           → reference to current object
│
└── ACCESS MODIFIERS
    ├── Public     → name      (open to all)
    ├── Protected  → _name     (convention: subclasses only)
    └── Private    → __name    (name-mangled, class only)
```
